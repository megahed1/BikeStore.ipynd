### 4] Data Analysis & Visualization (EDA & Statistical Analysis)
#### Exploratory Data Analysis

##### What is the average Revenue?

```python
average_revenue = df["revenue"].mean()
print(f"The average revenue is: {average_revenue}")
```

##### What were the highest and lowest Revenue?

```python
highst_revenue = df["revenue"].max()
lowest_revenue = df["revenue"].min()
print(f"The highest revenue is: {highst_revenue}")
print(f"The lowest revenue is: {lowest_revenue}")
```

##### How many orders have 'Road Bikes'  as a category name ?

```python
road_bikes_orders = df[df["category_name"] == "Road Bikes"]
print(f"The number of orders with 'Road Bikes' as a category name is: {len(road_bikes_orders)}")
```

##### Display the data which revenue greater than 10000

```python
highst_revenue = df[df["revenue"] > 10000]
print(highst_revenue)
```

##### what is the total revenue for each Brand ?

```python
revenue_per_brand = df.groupby("brand_name")["revenue"].sum()
revenue_per_brand
revenue_per_brand_sort = revenue_per_brand.sort_values(ascending=True)
revenue_per_brand_sort
```

```python
revenue_per_brand_sort.plot(kind="bar", title="Revenue per Brand")
```

##### what is the total revenue for each Category Name ?

```python
revenue_per_category = df.groupby("category_name")["revenue"].sum()
revenue_per_category
revenue_per_category_sort = revenue_per_category.sort_values(ascending=True)
revenue_per_category_sort
```

```python
revenue_per_category_sort.plot(kind="bar", title="Revenue per Category")
```
